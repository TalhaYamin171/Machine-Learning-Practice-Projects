```python
#Importing the necessary Libraries and the dataset for question 6 and 7.

import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

df = pd.read_csv("../input/iris/Iris.csv")

#Checking all the features to see if all are required
df.info()
```

The ID column is there just to maintain sequence and does not impact the target variable in anyway so it can be removed. 

```python
#Splitting the dataset into 80% train and 20% test as specified in question.
Train_Set, Test_Set = train_test_split(df,test_size=0.2)
```

```python
#Creating dataframes for training and testing with only the necessary features and target variable.
Indep_Train = Train_Set[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
Dep_Train = Train_Set[["Species"]]

Indep_Test = Test_Set[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
Dep_Test = Test_Set[["Species"]]
```

```python
#Training the model on the train set
KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(Indep_Train,Dep_Train.values.ravel())

#Testing Model on test Set
Dep_Pred=KNN.predict(Indep_Test)

#Calculating Accuracy
print('KNN accuracy = ',metrics.accuracy_score(Dep_Pred,Dep_Test))
```
